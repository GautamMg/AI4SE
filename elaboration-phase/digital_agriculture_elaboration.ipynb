Subsystem Overview:
* S1. Path Planner: Generates lawn-mower waypoints for the drone’s flight mission.
* S2. Edge Uploader: Monitors mission folder, batches new images, creates manifest, and uploads to HPC; triggers downstream jobs.
* S3. HPC Validator: Checks uploaded data integrity and compares EXIF metadata with planned parameters.
* S4. Compute Pipelines: Runs training/inference if validation passes; archives outputs.
* S5. Observability Layer: Logs system status, exposes APIs/dashboards for monitoring.

Target Environment:
* Platforms: Parrot Anafi (drone), Ubuntu Edge Device, Slurm-based HPC.
* Networking: Wi-Fi/LTE/VPN; uses HTTPS, Tapis, and optionally Globus.
* Tech Stack: Python (FastAPI), Tapis SDK, Docker, GitHub Actions.
* Security: Uses encrypted secrets and short-lived Tapis tokens.

Key Interfaces & Contracts:
* plan(...) → {waypoints, mission_manifest}
* submit(...) → job_uuid
* Uploader Config: Supports time/file/size batching.
* Validation Output: {status: "PASS/FAIL", reasons, metrics}

Prototype Goals & Acceptance Criteria

* P1. Path Planner: Correct spacing, aligned waypoints.

* P2. Edge Watcher: Recovers from network issues, prevents duplicate uploads, logs key stats.

* P3. Validator: Fast PASS/FAIL based on overlap/altitude; reasons clearly logged.

* P4. Tapis Trigger: Job UUID returned, status tracked, errors visible.

* P5. Observability: Batch transfer and job stats.
